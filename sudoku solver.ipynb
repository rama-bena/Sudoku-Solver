# Class SudokuSolver

```python
class SudokuSolver():
    def __init__(self):
        self.dxy = [(0,0), (1,0), (2,0),
            (0,1), (1,1), (2,1),
            (0,2), (1,2), (2,2)]
        self.is_solved = False
        self.hasil = pd.DataFrame()
        self.nol_terdikit = 81
        

    def solve(self, data):
        pq = self.buat_pq(data)
        self.rekursif(data, pq)

        if self.is_solved:
            return self.hasil
        else:
            print("tidak dapat diselesaikan")
            return self.hasil

    def rekursif(self, data, pq):
        if (self.is_solved):
            return
        
        banyak_nol = self.banyak_nol(data)
        if banyak_nol < self.nol_terdikit:
            self.nol_terdikit = banyak_nol
            self.hasil = data

        if banyak_nol == 0:
            self.is_solved = True
            return

        banyak, (x,y) = heapq.heappop(pq)
        kandidat = self.daftar_kandidat(data, x, y)

        for kand in kandidat:
            new_data = data.copy()
            new_data[x][y] = kand
            new_pq = self.buat_pq(new_data)
            self.rekursif(new_data, new_pq)

       
    def buat_pq(self, data):
        pq = []
        for x in range(9):
            for y in range(9):
                if data[x][y] == 0:
                    heapq.heappush(pq, (len(self.daftar_kandidat(data, x,y)), (x,y)))
        return pq


    def daftar_kandidat(self, data, x, y):
        kandidat = []
        for nilai in range(1, 10):
            if self.bisa_taruh(data, x, y, nilai):
                kandidat.append(nilai)
        return kandidat

    def banyak_nol(self, data):
        banyak = 0
        for x in range(9):
            for y in range(9):
                if data[x][y] == 0:
                    banyak += 1
        return banyak
        
    def bisa_taruh(self, data, x, y, nilai):
        x_mulai = (x//3) * 3
        y_mulai = (y//3) * 3

        for i in range(9):
            x_now, y_now = (x_mulai+self.dxy[i][0], y_mulai+self.dxy[i][1])
            if (data[x][i]==nilai) or (data[i][y]==nilai) or (data[x_now][y_now]==nilai):
                return False
        return True
```
